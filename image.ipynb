```python
#Usar Raw Strings:

#Inicie a string com um r antes das aspas simples para indicar uma raw string, onde as barras invertidas 
#so tratadas literalmente: Python
img = mpimg.imread(r'C:\Users\Edvaldojr\Pictures\stinkbug.png')
print(img)
```

```python
imgplot = plt.imshow(img)
```

```python
lum_img = img[:, :, 0]

# This is array slicing.  You can read more in the `Numpy tutorial
# <https://docs.scipy.org/doc/numpy/user/quickstart.html>`_.

plt.imshow(lum_img)
```

```python
plt.imshow(lum_img, cmap="hot")
```

```python
#Observe que voc tambm pode alterar mapas de cores em objetos de plotagem existentes usando o set_cmap()mtodo:
imgplot = plt.imshow(lum_img)
imgplot.set_cmap('nipy_spectral')
```

```python
# til ter uma ideia do valor que uma cor representa. Podemos fazer isso adicionando barras coloridas.
imgplot = plt.imshow(lum_img)
plt.colorbar()
```

```python
imgplot = plt.imshow(lum_img, clim=(0.0, 0.7))

```

```python
plt.hist(lum_img.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k')
```

```python
fig = plt.figure()
a = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(lum_img)
a.set_title('Before')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')
a = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(lum_img)
imgplot.set_clim(0.0, 0.7)
a.set_title('After')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')
```
